# A Data Science Framework

1. **Defina el problema:** Si la ciencia de los datos, el big data, el aprendizaje automático, el análisis predictivo, la inteligencia empresarial o cualquier otra palabra de moda es la solución, entonces. ¿cuál es el problema? Los problemas antes que los requisitos, los requisitos antes que las soluciones, las soluciones antes que el diseño y el diseño antes que la tecnología. **Con demasiada frecuencia nos apresuramos a saltar sobre la nueva y brillante tecnología, herramienta o algoritmo antes de determinar el problema real que estamos tratando de resolver.**
2. **Reunir los datos:** John Naisbitt escribió  que estamos "ahogados en datos, y sin embargo, faltos en conocimiento". Así que lo más probable es que el conjunto o conjuntos de datos ya existan en alguna parte, en algún formato. Como dice el refrán, no hay que reinventar la rueda, sólo hay que saber dónde encontrarla. En el siguiente paso, nos preocuparemos de transformar los "datos sucios" en "datos limpios".
3. **Preparar los datos para su consumo:** Este paso se suele denominar lucha de datos, un proceso necesario para convertir los datos "salvajes" en datos "manejables". El wrangling de datos incluye la implementación de arquitecturas de datos para el almacenamiento y el procesamiento, el desarrollo de estándares de gobernanza de datos para la calidad y el control, la extracción de datos (es decir, ETL y web scraping) y la limpieza de datos para identificar puntos de datos aberrantes, faltantes o atípicos.
4. **Perform Exploratory Data Analysis (EDA)**: Cualquiera que haya trabajado alguna vez con datos sabe que es garbage-in, garbage-out (GIGO). Por lo tanto, es importante desplegar estadísticas descriptivas y gráficas para buscar posibles problemas, patrones, clasificaciones, correlaciones y comparaciones en el conjunto de datos. Además, la categorización de los datos (es decir, cualitativa frente a cuantitativa) también es importante para comprender y seleccionar la prueba de hipótesis o el modelo de datos correctos.
5. **Model Data:** Al igual que las estadísticas descriptivas e inferenciales, el modelado de datos puede resumir los datos o predecir los resultados futuros. Su conjunto de datos y los resultados esperados, determinarán los algoritmos disponibles para su uso. Es importante recordar que los algoritmos son herramientas y no varitas mágicas o balas de plata. Un modelo erróneo puede llevar a un mal rendimiento en el mejor de los casos y a una conclusión equivocada en el peor caso.
6.  **Validate and Implement Data Model:** Después de haber entrenado su modelo basado en un subconjunto de sus datos, es el momento de probar su modelo. Esto ayuda a asegurar que no has sobreajustado tu modelo o lo has hecho tan específico para el subconjunto seleccionado, que no se ajusta con precisión a otro subconjunto del mismo conjunto de datos. En este paso determinamos si nuestro modelo se sobreajusta, generaliza o no se ajusta a nuestro conjunto de datos
7. **Optimize and Strategize:** este es el paso doden en el se repite el proceso para hacerlo mejor, más estable y más rápido que antes. 


Otra versión es la de Joseph F Hair en su libro "Multivariate Data Analysis", en la se desprenden los siguientes pasos:

1. **Entender el problema:** Veremos cada variable y haremos un análisis  sobre su significado e importancia para este problema.
2. **Estudio univariable:** Enfocada en la variable dependiente (Ej: Sobrevive o no) y de conocer un poco más sobre ella.
3. **Estudio multivariable:** Entender cómo se relacionan la variable dependiente y las variables independientes.
4. **Limpieza básica:** Limpiaremos el conjunto de datos y manejaremos los datos perdidos, los valores atípicos y las variables categóricas.
5. **Prueba de supuestos:** Comprobaremos si los datos cumplen los supuestos exigidos por la mayoría de las técnicas multivariantes.

## 3.21 Las 4 C's de la limpieza de datos (The 4 C's of Data Cleaning) : Correcting, Completing, Creating, and Converting

En esta etapa, limpiaremos nuestros datos 1) corrigiendo outliers o valores atípicos, 2) completando la información que falta (missing values), 3) creando nuevas características para el análisis y 4) convirtiendo los campos al formato correcto para los cálculos y la presentación.

1. **Correcting:** vemos que podemos tener posibles valores atípicos en la edad y la tarifa. Sin embargo, como son valores razonables, esperaremos hasta que completemos nuestro análisis exploratorio para determinar si debemos incluirlos o excluirlos del conjunto de datos. Cabe señalar que, si fueran valores poco razonables, por ejemplo, edad = 800 en lugar de 80, probablemente sea una decisión segura para arreglar ahora.
2. **Completing:** Hay valores nulos o datos que faltan en el campo edad, cabina y embarcado. Los valores perdidos pueden ser malos, porque algunos algoritmos no saben cómo manejar los valores nulos y fallarán. Mientras que otros, como los árboles de decisión, pueden manejar los valores nulos. Por lo tanto, es importante arreglar antes de empezar a modelar, porque vamos a comparar y contrastar varios modelos. Hay dos métodos comunes, eliminar el registro o rellenar el valor que falta utilizando una entrada razonable. No se recomienda eliminar el registro, especialmente un gran porcentaje de registros, a menos que realmente represente un registro incompleto. En su lugar, es mejor imputar los valores que faltan. Una metodología básica para los datos cualitativos es la imputación mediante el modo. Una metodología básica para los datos cuantitativos es imputar utilizando la media, la mediana o la media + la desviación estándar aleatoria. Una metodología intermedia es utilizar la metodología básica basada en criterios específicos; como la edad media por clase o el puerto de embarque por tarifa y SES. Existen metodologías más complejas, pero antes de desplegarlas hay que compararlas con el modelo base para determinar si la complejidad realmente añade valor. Para este conjunto de datos, la edad se imputará con la mediana, el atributo de cabina se eliminará y el embarque se imputará con la moda. Las iteraciones posteriores del modelo pueden modificar esta decisión para determinar si mejora la precisión del modelo.
3. **Creating:** Feature engineering es cuando utilizamos rasgos existentes para crear nuevos rasgos y determinar si proporcionan nueva i nformación para predecir nuestro resultado. 
4. **Converting:** Por último nos ocuparemos del formato. No hay formatos de fecha ni de moneda, sino de tipo de datos. Nuestros datos categóricos se importan como objetos, lo que dificulta los cálculos matemáticos. Para este conjunto de datos, convertiremos los tipos de datos objeto en variables ficticias categóricas.

```python
###COMPLETING: complete or delete missing values in train and test/validation dataset
for dataset in data_cleaner:    
    #complete missing age with median
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)

    #complete embarked with mode
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)

    #complete missing fare with median 
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)
    
#delete the cabin feature/column and others previously stated to exclude in train dataset
drop_column = ['PassengerId','Cabin', 'Ticket']
data1.drop(drop_column, axis=1, inplace = True)


print("Check 4C!")
print(data1.isnull().sum())
print("-"*10)
print(data_val.isnull().sum())
```

```python
###CREATE: Feature Engineering for train and test/validation dataset
for dataset in data_cleaner:    
    #Discrete variables
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

    dataset['IsAlone'] = 1 #initialize to yes/1 is alone
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0 # now update to no/0 if family size is greater than 1

    #quick and dirty code split title from name: http
    dataset['Title'] = dataset['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]


    #Continuous variable bins; qcut vs cut:
    """
    El comando cut crea contenedores equiespaciados pero la frecuenciade las muestras es desigual en cada contenedor
    El comando qcut crea contenedores de tamaño desigual, pero la frecuencia de las muestras es igual en cada contenedor.
    
    """
    #Fare Bins/Buckets using qcut or frequency bins: 
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4) #fare: valor del ticket

    #Age Bins/Buckets using cut or value bins
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)


    
#cleanup rare title names
print(data1['Title'].value_counts())
stat_min = 10 #while small is arbitrary, we'll use the common minimum in statistics
title_names = (data1['Title'].value_counts() < stat_min) #this will create a true false series with title name as index

#apply and lambda functions are quick and dirty code to find and replace with fewer lines of code
data1['Title'] = data1['Title'].apply(lambda x: 'Misc' if title_names.loc[x] == True else x)
print(data1['Title'].value_counts())
print("-"*10)


#preview data again
data1.info()
data_val.info()
data1.sample(10)
```

## 3.23 Convert Formats

Convertiremos los datos categóricos en variables dummy para el análisis. Hay múltiples formas de codificar las variables categóricas; nosotros utilizaremos las funciones de sklearn y pandas.

En este paso, también definiremos nuestras variables x (independiente/características/explicación/predictor/etc.) e y (dependiente/objetivo/resultado/respuesta/etc.) para el modelado de datos.


** Developer Documentation: **
* [Categorical Encoding](http://pbpython.com/categorical-encoding.html)
* [Sklearn LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)

```python
#CONVERT: convert objects to category using Label Encoder for train and test/validation dataset

#code categorical data
label = LabelEncoder()
for dataset in data_cleaner:    
    dataset['Sex_Code'] = label.fit_transform(dataset['Sex'])
    dataset['Embarked_Code'] = label.fit_transform(dataset['Embarked'])
    dataset['Title_Code'] = label.fit_transform(dataset['Title'])
    dataset['AgeBin_Code'] = label.fit_transform(dataset['AgeBin'])
    dataset['FareBin_Code'] = label.fit_transform(dataset['FareBin'])


#define y variable target/outcome
Target = ['Survived']

#define x variables for original features  feature selection
data1_x = ['Sex','Pclass', 'Embarked', 'Title','SibSp', 'Parch', 'Age', 'Fare', 'FamilySize', 'IsAlone'] #pretty name/values for charts
data1_x_calc = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code','SibSp', 'Parch', 'Age', 'Fare'] #coded for algorithm calculation
data1_xy =  Target + data1_x
print('Original X Y: ', data1_xy, '\n')


#define x variables for original w/bin features to remove continuous variables
data1_x_bin = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code', 'FamilySize', 'AgeBin_Code', 'FareBin_Code']
data1_xy_bin = Target + data1_x_bin
print('Bin X Y: ', data1_xy_bin, '\n')


#define x and y variables for dummy features original
data1_dummy = pd.get_dummies(data1[data1_x])
data1_x_dummy = data1_dummy.columns.tolist()
data1_xy_dummy = Target + data1_x_dummy
print('Dummy X Y: ', data1_xy_dummy, '\n')



data1_dummy.head()
```

# Paso 4: Perform Exploratory Data Analysis (EDA)

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(data1)
```
